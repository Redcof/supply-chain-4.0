```python

# display(Markdown('# *some markdown* $\phi$'))
markdown = lambda x: display(Markdown(x))
```

```python
def series_2_wordcloud(ax, df, dataset, col):
    text = " ".join(i for i in df[col])
    stopwords = set(STOPWORDS)
    wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text)
    ax.set_title("%s:[%s]"%(dataset, col), fontdict=dict(fontweight="bold"))
    ax.imshow(wordcloud, interpolation='bilinear')
```

## Frequency and Wordcoud

```python
%time
path = r"/Users/soumensardar/Library/CloudStorage/GoogleDrive-soumensardarintmain@gmail.com/.shortcut-targets-by-id/1zlA8VX-l_IufCZSfJIfxohpsEGrcLx4-/Taniya Paul Thesis paper/writing/Figures and Charts"
for dataset_name in ["online_retail", "online_retail_2", "product_demand", "food_demand", "livestock_meat_import"][-1:]:
    df, target, timeseries_col, dataset_name = get_dataset(dataset_name, ablation_limit=ablation,
                                                               is_extra_feature_enabled=is_extra_feature_enabled, label_encoding=False)
    markdown("<hr/><hr/></hr/></hr/>")
    markdown("# `%s` Dataset"%dataset_name)
    cols = [col for col in df.columns if df[col].dtype in ("string", "object")]
    n_cols = len(cols)
    if n_cols == 0:
        continue
    n_rows = 1
    n_items = 3
    if n_cols > n_items:
        n_rows = (n_cols // n_items) + 1
    print(n_rows, n_cols)
    plt.figure(figsize=((n_cols*3)+1, 4*n_rows))
    plt.axis("off")
    for idx, col in enumerate(cols):
        ax = plt.subplot(n_rows, n_items, idx+1)
        if df[col].nunique() > 10:
            series_2_wordcloud(ax, df, dataset_name, col)
        else:
            ax.set_title("%s[%s]"%(dataset_name, col), fontdict=dict(fontweight="bold"))
            df[col].value_counts(normalize=True).plot.pie(ax=ax, autopct='%1.1f%%')
            ax.set_ylabel("")
    # plt.savefig(f"{path}/{dataset_name}-eda.png")
    plt.show()
    %time
```

## Auto EDA

```python
%time
for dataset_name in ["online_retail", "online_retail_2", "product_demand", "food_demand", "livestock_meat_import"]:
    df, target, timeseries_col, dataset_name = get_dataset(dataset_name, ablation_limit=ablation,
                                                               is_extra_feature_enabled=is_extra_feature_enabled, label_encoding=False)
    markdown("<hr/><hr/></hr/></hr/>")
    markdown("# `%s` Dataset"%dataset_name)
    config = dict(
        correlations = {"pearson" : {"calculate" : True}, "spearman":{"calculate": True}},
        missing_diagrams={"heatmap": False, "matrix": False, "bar": False},
        interactions={"targets":[target]},
        duplicates=None,
        samples=None,
    )
    display(ProfileReport(df, title="%s-Profiling Report"%dataset_name, tsmode=False, sortby=timeseries_col, **config))
    report = sweetviz.analyze(df, target_feat=target)
    display(report.show_notebook())
    %time
```

```python
%time
import matplotlib.pyplot as plt
%matplotlib inline
for dataset_name in ["online_retail", "online_retail_2", "product_demand", "food_demand", "livestock_meat_import"]:
    df, target, timeseries_col, dataset_name = get_dataset(dataset_name, ablation_limit=ablation,
                                                               is_extra_feature_enabled=is_extra_feature_enabled, label_encoding=False)
    markdown("<hr/><hr/></hr/></hr/>")
    markdown("# `%s` Dataset"%dataset_name)
    config = dict(
        correlations = {"pearson" : {"calculate" : True}, "spearman":{"calculate": True}},
        missing_diagrams={"heatmap": False, "matrix": False, "bar": False},
        interactions={"targets":[target]},
        duplicates=None,
        samples=None,
    )
    display(ProfileReport(df, title="%s-Profiling Report"%dataset_name, tsmode=False, **config))
    report = sweetviz.analyze(df, target_feat=target)
    display(report.show_notebook())
    if timeseries_col:
        df[[target,	"lag_1",	"lag_2",	"lag_3",	"rolling_mean",	"fourier_transform", timeseries_col]].plot.line(x=timeseries_col, title=dataset_name, subplots=True, figsize=(20,15))
    else:
        df[[target,	"lag_1",	"lag_2",	"lag_3",	"rolling_mean",	"fourier_transform"]].plot.line(title=dataset_name, subplots=True, figsize=(20,15))
    plt.show()
    %time
```
